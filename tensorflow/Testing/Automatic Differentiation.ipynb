```python
%matplotlib inline
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
from d2l import tensorflow as d2l
from IPython import display
```

```python
x = tf.range(4, dtype=tf.float32)
x
```

```python
x = tf.Variable(x)
```

```python
with tf.GradientTape() as t:
    y = 2*tf.tensordot(x,x,axes=1)
y
```

```python
x_grad = t.gradient(y,x)
x_grad
```

```python
x_grad == 4*x
```

```python
with tf.GradientTape() as t:
    y = tf.reduce_sum(x)
t.gradient(y, x)
```

```python
with tf.GradientTape(persistent=True) as t:
    y = x * x
    u = tf.stop_gradient(y)
    z = u * x

x_grad = t.gradient(z, x)
x_grad == u
```

```python
t.gradient(y, x) == 2*x
```

```python
def f(a):
    b=a*2
    while tf.norm(b) < 1000:
        b = b *2
    if tf.reduce_sum(b) >0:
        c =b
    else:
        c = 100 *b
    return c
```

```python
a = tf.Variable(tf.random.normal(shape=()))
with tf.GradientTape() as t:
    d=f(a)
d_grad = t.gradient(d, a)
d_grad
```

```python
d_grad == d/a
```

```python
def use_svg_display(): #@save
    display.set_matplotlib_formats('svg')
```

```python
def set_figsize(figsize=(3.5,2.5)): #@save
    use_svg_display()
    d2l.plt.rcParams['figure.figsize'] = figsize
```

```python
def set_axes(axes, xlabel,ylabel,xlim,ylim,xscale,yscale,legend):
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.set_xscale(xscale)
    axes.set_yscale(yscale)
    axes.set_ylim(ylim)
    axes.set_xlim(xlim)
    if legend:
        axes.legend(legend)
    axes.grid()
```

```python
def plot(X,Y=None, xlabel=None, ylabel=None, legend=None, xlim=None, ylim=None,
         xscale='linear', yscale='linear', fmts=('-', 'm--', 'g-.','r:'), figsize=(3.5, 2.5), axes=None):
    if legend is None:
        legend = []
    set_figsize(figsize)
    axes = axes if axes else d2l.plt.gca()
    
    def has_one_axis(X):
        return (hasattr(X, "ndim") and X.ndim == 1 or isinstance(X, list) and not hasattr(X[0], "__len__"))
    if has_one_axis(X):
        X=[X]
    if Y is None:
        X,Y = [[]] * len(X), X
    elif has_one_axis(Y):
        Y=[Y]
    if len(X) != len(Y):
        X = X * len(Y)
    axes.cla()
    for x,y, fmt in zip(X,Y,fmts):
        if len(x):
            axes.plot(x,y,fmt)
        else:
            axes.plot(y,fmt)
    set_axes(axes, xlabel, ylabel, xlim, ylim, xscale, yscale, legend)
```

```python
x = tf.range(-10, 10, 0.1)
x = tf.Variable(x)

with tf.GradientTape() as t:
    y = tf.math.sin(x)

x_grad = t.gradient(y, x)

x = np.arange(-10,10,0.1)
plt.figure(1)
plt.plot(x,np.sin(x),color='r')
plt.plot(x,x_grad.numpy(), color='g')
plt.show()
```
